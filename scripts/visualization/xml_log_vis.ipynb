```python
resolution = 1000
my_dpi = 200

show_avg_speed_label = False

# fig, ax = plt.subplots(2, 1, figsize=(resolution/my_dpi, 1.2 *resolution/my_dpi), dpi=my_dpi)

fig, ax = plt.subplots(nrows=2, ncols=1, gridspec_kw={'height_ratios':[6,2]}, figsize=(resolution/my_dpi, 8/6 *resolution/my_dpi), dpi=my_dpi)

ax2 = ax[1]
ax = ax[0]

width =  6.0
height = 6.0



ax.grid(True)
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
ax.tick_params(axis="y",direction="in", pad=-20)
ax.tick_params(axis="x",direction="in", pad=-20)
# ax.xaxis.set_ticks(np.arange(0.0, 40.0, 1.0))
# ax.yaxis.set_ticks(np.arange(0.0, 40.0, 1.0))
# ax.set_xlim((0, 20))
# ax.set_ylim((0, 20))
ax.axis('equal')

colors = [name for name, color in mcolors.TABLEAU_COLORS.items()]
colors = colors[:5]
style = "Simple, tail_width=0.2, head_width=1, head_length=2"


print(default_size)
for a in range(agents_num):
    circle = plt.Circle(steps_log[-1, a, 0:2], default_size, color=colors[a%len(colors)], alpha=0.8)
    ax.add_patch(circle)

for a in range(agents_num):
    ax.plot(steps_log[:, a, 0], steps_log[:, a, 1], lw=2, color=colors[a%len(colors)])


velocities = []
all_velocities = []
for step in range(len(steps_log)-1):
    curr_avg_vel = 0
    for a in range(agents_num):
        curr_avg_vel += (np.linalg.norm(steps_log[step+1, a, 0:2]-steps_log[step, a, 0:2]) / 0.1) 
        all_velocities.append(np.linalg.norm(steps_log[step+1, a, 0:2]-steps_log[step, a, 0:2]) / 0.1)
    curr_avg_vel /= agents_num
    velocities.append(curr_avg_vel)
if show_avg_speed_label:  
    ax.set_ylabel("Y")
ax.set_xlabel("X")


time = np.linspace(0, (len(steps_log)-1)*0.1, (len(steps_log)-1))
ax2.plot(time, velocities)
mean_vel = np.array(all_velocities).mean()
print(mean_vel)
ax2.axhline(mean_vel, c='tab:red')

ax2.set_ylim((0, 1.0))
ax2.set_xlim((0, (len(steps_log)-1)*0.1))
ax2.yaxis.set_ticks(np.arange(0, 1.1, 0.2))

if not show_avg_speed_label:
    ax2.axes.yaxis.set_ticklabels([])
    ax2.tick_params(axis="y",direction="in", pad=-20)

ax2.xaxis.set_ticks(np.arange(0, (len(steps_log)-1)*0.1, 5.0))

ax2.grid( linestyle=':')
if show_avg_speed_label:    
    ax2.set_ylabel("Avg. Speed, m/s")
ax2.set_xlabel("Time, s")

# ax2.yaxis.grid()
fig.tight_layout()
fig.savefig(image_name + image_ext, transparent=False, pad_inches=0.0001, dpi=my_dpi)
```

```python


colors =  ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
trcolors =  ['#1f77b4c8', '#ff7f0ec8', '#2ca02cc8', '#d62728c8', '#9467bdc8', '#8c564bc8', '#e377c2c8', '#7f7f7fc8', '#bcbd22c8', '#17becfc8']

colors = colors[:5]
colors = [colors[a%len(colors)] for a in range(agents_num)]
trcolors = trcolors[:5]
trcolors = [trcolors[a%len(trcolors)] for a in range(agents_num)]
WHITE = (255, 255, 255)
step = 0
images = []


min_x = 0
min_y = 0
max_x = 20
max_y = 20
points_per_unit = 40
offset = np.array([min_x, min_y])

resolution_x = (max_x - min_x) * points_per_unit
resolution_y = (max_y - min_y) * points_per_unit

points_per_unit *= 2
resolution_x *= 2
resolution_y *= 2

agent_radius = default_size * points_per_unit
grid_step = 1.0
grid_x = np.arange(0, resolution_x, grid_step * points_per_unit)
grid_y = np.arange(0, resolution_y, grid_step * points_per_unit)

line_width = int(0.07 * points_per_unit)
circle_outline = int(0.05 * points_per_unit)

base_im = Image.new('RGB', (resolution_x, resolution_y), WHITE)
draw = ImageDraw.Draw(base_im, 'RGBA')

for x in grid_x: 
    draw.line(((x, 0), (x, resolution_y)), width=4, joint='curve', fill=(200,200,200))
for y in grid_y: 
    draw.line(((0, y), (resolution_x, y)), width=4, joint='curve', fill=(200,200,200))

del draw

for time_step in range(steps_num):

    im = base_im.copy()
    draw = ImageDraw.Draw(im, 'RGBA')

    for a in range(agents_num):
        path = []
        for pos in steps_log[:time_step+1, a, 0:2]:
            pos = pos - offset
            path.append((pos[0] * points_per_unit, pos[1] * points_per_unit))
        draw.line(path, width=line_width, joint='curve', fill=colors[(a)%len(colors)])
        


    for a in range(agents_num):
        a_pos = (steps_log[time_step, a, 0:2] - offset) * points_per_unit
        draw.ellipse ((a_pos[0] - agent_radius, a_pos[1] - agent_radius, a_pos[0]+agent_radius, a_pos[1]+agent_radius), fill=trcolors[(a)%len(trcolors)], outline=colors[(a)%len(colors)], width=circle_outline)

        # if time_step > 0:
        #     vxvy = steps_log[time_step, a, 0:2] - steps_log[time_step-1, a, 0:2] 
        #     vxvy = (vxvy / (np.linalg.norm(vxvy) + 0.000001)) * (default_size + 0.05)
        #     # print(np.linalg.norm(vxvy))
        #     ec_pos = (vxvy + steps_log[time_step, a, 0:2] - offset) * points_per_unit
        #     er = (agent_radius + 0.05) * 2
        #     # print(ec_pos)
        #     draw.ellipse ((ec_pos[0] - er, ec_pos[1] - er, ec_pos[0]+er, ec_pos[1]+er), fill=None, outline=colors[(a)%len(colors)], width=circle_outline)



    for a in range(agents_num):
        g_pos = (goal_states[a, 0:2] - offset) * points_per_unit
        draw.rectangle ((g_pos[0] - agent_radius, g_pos[1] - agent_radius, g_pos[0]+agent_radius, g_pos[1]+agent_radius), fill=trcolors[(a)%len(trcolors)], outline=colors[(a)%len(colors)], width=circle_outline)    
    
    im = ImageOps.flip(im)
    fnt = ImageFont.truetype("/usr/share/fonts/noto/NotoSans-BoldItalic.ttf", 30)
        # fnt = ImageFont.load_default()
    draw = ImageDraw.Draw(im, 'RGBA')
    draw.text((30,20), f"Simulation Time : {int(time_step * 0.1)//60:02}:{int(time_step * 0.1)%60:02}:{int(time_step) %10 *10:02}", font=fnt, fill=(0,0,0))

    im = im.resize((resolution_x // 2, resolution_y // 2), resample=Image.LANCZOS)
    
    
    images.append(im)
    
plt.close(fig)

images[0].save(animation_name+animation_ext, save_all=True, append_images=images[1:], optimize=False, duration=100, loop=0)
display(Img(filename = animation_name+animation_ext, width = 800, height = 800))
```
